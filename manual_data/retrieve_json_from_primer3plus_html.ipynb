```python
from bs4 import BeautifulSoup
import json
from pprint import pprint
```

```python
source_html_file = open("./primer3plus-default.html", "r")
source_html = source_html_file.read()
source_html_file.close()
soup = BeautifulSoup(source_html, "html.parser")
```

```python
print(soup.prettify())
```

```python
data = soup.findAll("input")
print(data[0])
```

```python
data = {}
for tag in soup.find_all("input") :
    if tag.get('name') is not None:
        data[tag.get('name')] = tag.get('value')

for tag in soup.find_all("textarea") :
    if tag.get('name') is not None:
        data[tag.get('name')] = tag.get('value')

for tag in soup.find_all("select") :
    if tag.get('name') is not None:
        data[tag.get('name')] = tag.get('value')

print(len(data.keys()))
```

```python
from collections import OrderedDict
orderedData = OrderedDict(sorted(data.items()))
pprint(orderedData)
```
